```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x and y coordinates
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)

# Create a grid of points
X, Y = np.meshgrid(x, y)

# Define the angle, frequency, and phase
angle = 1.3*np.pi / 4  # 45 degrees
frequency = 10
phase = 0

# Calculate the z values at each point in the grid
Z = np.sin(frequency * (X * np.cos(angle) + Y * np.sin(angle)) + phase)

# Display the 2D sin wave as an image
plt.imshow(Z, cmap='viridis', origin='lower', extent=[-6, 6, -6, 6])

# Show the plot
plt.show()
```

```python
# Compute the 2D FFT of the sin wave image
Z_fft = np.fft.fft2(Z)

# Shift the zero-frequency component to the center of the spectrum
Z_fft_shifted = np.fft.fftshift(Z_fft)

# Display the absolute value of the FFT result as an image
plt.imshow(np.abs(Z_fft_shifted), cmap='viridis', origin='lower', extent=[-6, 6, -6, 6])

# Show the plot
plt.show()
```

```python
import numpy as np

zoomfact = 1.0

ny=512
nx=512
fact=1.0
xcent=nx//2
k0x=1.0
ycent=ny//2
k0y=0.5

dxy = 0.1
coslookup = np.zeros((int(zoomfact*ny), int(zoomfact*nx)))
sinlookup = np.zeros((int(zoomfact*ny), int(zoomfact*nx)))

for i in range(int(zoomfact*ny)):
    for j in range(int(zoomfact*nx)):
        angle = fact * np.pi * ((j-xcent)*k0x + (i-ycent)*k0y) * dxy/zoomfact
        coslookup[i, j] = np.cos(angle)
        sinlookup[i, j] = np.sin(angle)
```
